#### （a）现在将深度分为七个等级：[0,5,10,15,20,30,50,np.inf]，请以深度等级Ⅰ,Ⅱ,Ⅲ,Ⅳ,Ⅴ,Ⅵ,Ⅶ为索引并按照由浅到深的顺序进行排序。

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/Earthquake.csv')
df.head()
```

```python
s = pd.cut(df['烈度'], [0,5,10,15,20,30,50,np.inf], include_lowest=True, labels=['Ⅰ','Ⅱ','Ⅲ','Ⅳ','Ⅴ','Ⅵ','Ⅶ'])
df.set_index(s).sort_index().head()
```
